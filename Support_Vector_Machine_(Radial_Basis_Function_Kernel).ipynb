```python
# pd.set_option('display.float_format', lambda x: f'%.{len(str(x%1))-2}f' % x)
df = pd.read_excel("Bdiabetes.xlsx")
#df = pd.read_csv("Pima indian.csv")
df
```

# TRAIN TEST SPLIT DATA

```python
# Generate some sample data
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values


# Define the number of folds
#k = 2

# Shuffle the data
#indices = np.random.permutation(len(X))

#Split the data into k folds
#folds = np.array_split(indices, k)

 #  k-fold cross-validation
#for i in range(k):
     # Select the test and training data for this fold
    #test_indices = folds[i]
    #train_indices = np.concatenate([f for j, f in enumerate(folds) if j != i])
    #X_train, y_train = X[train_indices], y[train_indices]
    #X_test, y_test = X[test_indices], y[test_indices]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=0)

print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

# NORMALISASI DATA TRAINING

```python
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))


# Define the minimum and maximum values for scaling
min_val = np.min(X_train)
max_val = np.max(X_train)

# Apply min-max scaling
norm_training = (X_train - min_val) / (max_val - min_val)

print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

# KERNEL RBF DATA TRAINING

```python
sigma = 0.001

def rbf(norm_training):
  K = np.zeros((len(norm_training), len(norm_training)))
  for i in range (len(norm_training)):
    for j in range (len(norm_training)):
      #K[i,j] = math.exp(-omega*np.linalg.norm(norm_training[i] - norm_training[j])**2)
      K[i,j] = math.exp(-(np.linalg.norm(((norm_training[i] - norm_training[j])**2)/(2*(sigma**2)))))
  return K

svm = rbf(norm_training)
print(svm)
```

# MATRIKS HESSIAN

```python
ytn=np.zeros(len(y_train))

for y in range(len(y_train)) :
    if y_train[y] == 0 :
        ytn[y] = -1
    else :
        ytn[y] = 1
#print(len(ytn))
#print(y_train)
#print(ytn)
```

```python
#print(norm_training)
#print(y_train)
#print(X_test)
#print(y_test)
hessian = np.zeros((len(y_train), len(y_train)))
print(len(hessian))
lambeda = 1
for o in range(len(y_train)):
  for p in range(len(y_train)):
    #if y_train[p] == 0:
    #hessian[o][p] = 1*-1*(svm[o][p]+(lambeda**2))
    hessian[o][p] = ytn[o]*ytn[p]*(svm[o][p]+(lambeda**2))
    #if y_train[p] == 1:
        #hessian[o][p] = 1*1*(svm[o][p]+(lambeda**2))
print(hessian)
```

# SEQUENTIAL TRAINING

```python
iter = 100
t = 0
C = 1
gamma = 0.0001
eps = 0.001
ai = [0]*len(y_train)
dai = [0]*len(y_train)
daitemp = 0
eip = [0]*len(y_train)
ei = np.zeros((len(y_train), len(y_train)))
while t < iter :
  for r in range(len(y_train)):
    for e in range(len(y_train)):
      ei[r][e] = hessian[r][e]*gamma
    eip[r] = np.sum(ei[r][e])
    daitemp = min(max((gamma*(1-eip[r])),ai[r]), C-ai[r])
    if daitemp <= eps:
      dai[r] = daitemp
      ai[r] = ai[r]+dai[r]
    else:
      r = len(y_train)
      t = iter
  t = t + 1
print(dai)
print(ai)
#print(eip)
```

# WX+ WX-

```python
# rows, cols = (len(y_train), 2)
# positip = [[0]*cols]*rows
# negatip = [[0]*cols]*rows
positip = [0]*len(y_train)
negatip = [0]*len(y_train)
for posneg in range(len(y_train)):
  if y_train[posneg] == 1:
    positip[posneg] = ai[posneg]
    # positip[posneg][1] = y_train[posneg]
  else :
    negatip[posneg] = ai[posneg]
    # negatip[posneg][1] = y_train[posneg]
positip = [h1 for h1 in positip if h1 is not 0]
negatip = [h2 for h2 in negatip if h2 is not 0]
wxplus = max(positip)
wxmin = max(negatip)
#print(positip)
#print(negatip)
print(wxplus)
print(wxmin)
indwxplus = positip.index(wxplus)
indwxmin = negatip.index(wxmin)
print(indwxplus)
print(indwxmin)
```

# INDEX KERNEL WX+ WX-

```python
indrbfplus = svm[indwxplus]
indrbfmin = svm[indwxmin]
print(indrbfplus)
print(indrbfmin)
```

# TOTAL HASIL WX+ WX-

```python
h_wxplus = [0]*len(y_train)
h_wxmin = [0]*len(y_train)
for t_wx in range(len(y_train)):
  if y_train[t_wx] == 1:
    h_wxplus[t_wx] = ai[t_wx]*1*indrbfplus[t_wx]
  else :
    h_wxmin[t_wx] = ai[t_wx]*-1*indrbfmin[t_wx]

h_wxplus = [h3 for h3 in h_wxplus if h3 is not 0]
h_wxmin  = [h4 for h4 in h_wxmin if h4 is not 0]

print(h_wxplus)
print(h_wxmin)
#print(ai)
```

# MENGHITUNG NILAI BIAS

```python
t_wxplus = sum(h_wxplus)
t_wxmin  = sum(h_wxmin)

bias = -0.5*(t_wxplus + t_wxmin)

print(t_wxplus)
print(t_wxmin)
print(bias)
```

# CONFUSION MATRIKS

```python
confus = [[0,0],[0,0]]

for f in range(len(y_test)):
  if y_test[f] == 1 and outcome[f] == 1:
    confus[1][1] +=1
  elif y_test[f] == 1 and outcome[f] == 0:
    confus[1][0] +=1
  elif y_test[f] == 0 and outcome[f] == 1:
    confus[0][1] +=1
  elif y_test[f] == 0 and outcome[f] == 0:
    confus[0][0] +=1

tn = confus[0][0]
fn = confus[1][0]
fp = confus[0][1]
tp = confus[1][1]

accuracy = (tn + tp) / (tn+tp+fn+fp)
precision = tp / (tp+fp)
recall = tp / (tp+fn)
f1_score = (2*precision*recall) / (precision+recall)

print(accuracy)
print(precision)
print(recall)
print(f1_score)
print(y_test)
print(outcome)
print(confus[0][0])
print(confus[1][0])
print(confus[0][1])
print(confus[1][1])
```
